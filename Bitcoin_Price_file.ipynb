```python
# Dependencies
import pandas as pd
import numpy as np
import quandl
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
# download bitcoin price
bitcoin = quandl.get("BCHAIN/MKPRU")
```

```python
# data clean
bitcoin.columns = ['Close']
bitcoin['Close'].replace(0, np.nan, inplace=True)
bitcoin = bitcoin.dropna()
```

```python
# plot bitcoin price
bitcoin.Close.plot(logy=True);
plt.xlim(pd.Timestamp('2013-12-01'), pd.Timestamp('2014-12-01'))
plt.ylim(300, 1000)
plt.legend()
plt.show()
```

```python
# plot bitcoin price of last 12 months
bitcoin.Close.plot(logy=True);
plt.xlim(pd.Timestamp('2017-03-15'), pd.Timestamp('2018-03-15'))
plt.ylim(900, 22000)
plt.legend()
plt.show()
```

```python
# reset index
bitcoin.reset_index(inplace=True)
```

```python
# check final data frame
bitcoin.head()
bitcoin.tail()
```

```python
# save CSV file
bitcoin.to_csv("bitcoinPrice.csv")
```
